##### 2. You have arr = np.array([50, 30, 40, 20]) and names = np.array(["A", "B", "C", "D"]).
##### Sort the arr values in ascending order and rearrange the names array accordingly.

```python
def Sort_Array(arr,names):
    sorted_indices_arr = np.argsort(arr)
    sorted_arr = arr[sorted_indices_arr]
    sorted_names = names[sorted_indices_arr]
    print("Sorted arr:", sorted_arr)
    print("Rearranged names:", sorted_names)
arr = np.array([50,30,40,20])
names = np.array(['A','B','C','D'])
Sort_Array(arr,names)
```

##### 3. Write code to find the k smallest elements from the array arr. Print these elements in ascending order.

```python
def k_smallest_elements(arr,k):
    Partitioned_Array = np.partition(arr,k)
    small_array = Partitioned_Array[:k]
    sorted_elements = np.sort(small_array)
    print(f'{k} smallest elements from the array arr are {sorted_elements}')
arr = np.array([1,21,46,928,44,90,36,56,430])
k = 4
k_smallest_elements(arr,k)
```

##### 6. Print k largest elements of an array arr in descending order.(use np.partition() to find top k elements)

```python
arr = np.array([5,6,2,8,1,3,9,4,0,7])
k = 4
partitioned_array = np.partition(arr, -k)
k_largest_elements = partitioned_array[-k:]
largest_elements_sorted = np.sort(k_largest_elements)[::-1]
print(f"The {k} largest elements in descending order are {largest_elements_sorted}")
```

##### 7. Given an array arr(containing possible negative values), sort the array in ascending order based on the absolute values of the elements. Print the sorted array.(use np.argsort() with a custom key for absolute values)

```python
arr = np.array([5, -6, 2, 8, -1, 3, -9, 4, 0, -7])
sorted_arr = arr[np.argsort(np.abs(arr))]
print("Array sorted by absolute values of negative elements:", sorted_arr)
```

##### 8. Given an array arr, sort only the elements greater than k in ascending order, keeping the other elements in their original positions.(create a mask to find elements greater than k, use np.sort() to sort them then reassign the sorted values back to their original indices in the array)

```python
arr = np.array([5, 8, 2, 12, 1, 7, 9, 4, 10])
k = 7
mask = arr > k
sorted_elements = np.sort(arr[mask])
arr[mask] = sorted_elements
print(f'Array after sorting elements greater than k is {arr}')
```

### Section 4: Indexing and slicing

##### 1. 2D Array Indexing 

```python
arr_2d = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
print(f"Accessing element at row 1 and column 2 is {arr_2d[1,2]}")
first_row = arr_2d[0,:]
print(f"First row is {first_row}")
last_column = arr_2d[:,-1]
print(f'The last column is {last_column}')
```

##### 2. Boolean Indexing

```python
arr = np.array([2, 5, 8, 11, 14, 17, 20])
greater_than_10 = arr[arr>10]
print(f'Values greater than 10 are {greater_than_10}')
arr[arr>15] = -1
print(f'After replacing values greater than 15 with -1, Array will be {arr}')
```

##### 3. Fancy Indexing

```python
arr_2d = np.array([[5, 10, 15], [20, 25, 30], [35, 40, 45]])
row_array = [0,1,2]
column_array = [2,1,0]
select_elements = arr_2d[row_array, column_array]
print("Selected elements:", select_elements)
submatrix=arr_2d[np.array([0,2])[:,np.newaxis],np.array([1,2])]
print(f'Submatrix: \n{submatrix}')
```

##### 4. Slicing 1D array

```python
arr = np.array([1, 3, 5, 7, 9, 11, 13, 15])
first_4_elements = arr[:4]
every_second_element = arr[::2]
elements_from_2_to_end_step_2 = arr[2::2]
print("First 4 elements:", first_4_elements)
print("Every second element:", every_second_element)
print("Elements from index 2 to the end, stepping by 2:", elements_from_2_to_end_step_2)
```

##### 6. Handling Out-of-Range Indices in Slicing vs. Indexing

```python
arr = np.array([10, 20, 30, 40, 50])
print("The output of the following slicing operations is:")
print(arr[2:10])
print(arr[-10:3])
```

##### Slicing will not raise an error for out-of-range indices. It will simply return as many elements as are available within the array bounds, up to the specified limit (here index 10). If the end index exceeds the length of the array, it will return all available elements from the start index to the end of the array.

##### 7. Work with an Audio File

```python
audio_file_path = '/home/abhinav/Downloads/simple-logo-149190.mp3'
audio = AudioSegment.from_mp3(audio_file_path)
samples = np.array(audio.get_array_of_samples())
frame_rate = audio.frame_rate

first_second_samples = samples[:frame_rate]

last_second_samples = samples[-frame_rate:]

combined_samples = np.concatenate((first_second_samples, last_second_samples))

# Plot the waveforms
plt.figure(figsize=(10, 6))

# Plot first second waveform
plt.subplot(3, 1, 1)
plt.plot(np.linspace(0, 1, frame_rate), first_second_samples)
plt.title("First Second of the Audio")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")

# Plot last second waveform
plt.subplot(3, 1, 2)
plt.plot(np.linspace(0, 1, frame_rate), last_second_samples)
plt.title("Last Second of the Audio")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")

# Plot combined waveform
plt.subplot(3, 1, 3)
plt.plot(np.linspace(0, 2, 2 * frame_rate), combined_samples)
plt.title("Combined First and Last Second of the Audio")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")

# Display the plots
plt.tight_layout()
plt.show()
```
